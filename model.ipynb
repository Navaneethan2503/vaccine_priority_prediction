```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
import pandas as pd
df = pd.read_csv('DATA_NO.csv')
df['male']=df['Gender']=='M'
X=df[['Current','male','Age','Num Cases', 'Type_of_worker', 'Problem']].values
y=df['Target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=22);
model.fit(X_train, y_train)
```

```python
print(model.predict([[3,True,35,1,3,1]]))
```

```python
print('Accuracy:',model.score(X_test, y_test))
```

```python
print(model.predict([[3,True,30,1,3,0]]))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.tree import export_graphviz
feature_names = ['Current','male','Age','Num Cases', 'Type_of_worker', 'Problem']
X= df[feature_names].values
y=df['Target'].values
dt = DecisionTreeClassifier()
dt.fit(X, y)
import graphviz
dot_file = export_graphviz(dt, feature_names=feature_names)
graph = graphviz.Source(dot_file)
graph.render(filename='covid_tree', format='png', cleanup=True)

```
